#**Hate Speech Detection**

```python
df=pd.read_csv("/content/HateSpeechData.csv")
df
```

```python
df["labels"]=df["class"].map({0:"Hate Speech Detected",1:"Offensive Language detected",2:"No hate and offensive speech"})
df
```

```python
df=df[["tweet","labels"]]
df
```

```python
def clean(text):
  text=str(text).lower()
  text=re.sub('\[.*?]','',text)
  text=re.sub('https?://\S+|www\.\S+','',text)
  text=re.sub('<.*?>+','',text)
  text=re.sub('\n','',text)
  text=text.translate(str.maketrans('', '', string.punctuation))
  text=re.sub('\w*\d\w*','',text)
  text=[word for word in text.split(' ') if word not in stopword]
  text=" ".join(text)
  return text
df["tweet"]=df["tweet"].apply(clean)
df
```

```python
x=np.array(df["tweet"])
y=np.array(df["labels"])

cv=CountVectorizer()
x=cv.fit_transform(x)

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)


clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#**Gender Detection**

```python
import pandas as pd
sdf=pd.read_csv("gender-classifier-DFE-791531.csv",encoding="latin")
sdf
```

```python
data1=sdf[["_unit_id","created","name","gender","text","description"]]
data1
```

```python
data1["gender"].value_counts()
```

```python
data1.dropna(subset=["gender"], inplace=True)
```

```python
data1.drop(data1.loc[data1["gender"]=="unknown"].index, inplace=True)
data1.drop(data1.loc[data1["gender"]=="brand"].index, inplace=True)
```

```python
data1["gender"].value_counts()
```

```python
data1.dropna(subset=["description"], inplace=True)
```

```python
data1.reset_index(inplace=True)
data1
```

```python
data1["tweetdesc"]=data1["text"]+" "+data1["description"]
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

# Define the preprocessing function
def preprocess_text(text):
    # 1. Convert text to lowercase
    text = text.lower()

    # 2. Remove special, non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)

    # 3. Remove stopwords using NLTK
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    text = ' '.join([word for word in word_tokens if word not in stop_words])

    # 4. Remove emoticons, retweets, favorites, hashtags, URLs, and usernames starting with '@'
    text = re.sub(r'@\w+\s?', '', text)  # Remove usernames starting with '@'
    text = re.sub(r'@(\w+)', '', text)
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'#\w+', '', text)  # Remove hashtags
    text = re.sub(r'RT\s', '', text)  # Remove retweets
    text = re.sub(r'&\w+;', '', text)  # Remove HTML entities
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 5. Remove duplicates
    text = ' '.join(dict.fromkeys(text.split()))
    # 6. Tokenize using NLTK
    tokens = word_tokenize(text)

    return ' '.join(tokens)
```

```python
data1['ctext'] = data1['text'].apply(preprocess_text)
```

```python
data1['cdescription'] = data1['description'].apply(preprocess_text)
```

```python
data1['ctweetdesc'] = data1['tweetdesc'].apply(preprocess_text)
```

```python
data1.dropna(subset=["cdescription"], inplace=True)
data1.reset_index(inplace=True)
data1
```

```python
data1['numgender'] = data1['gender'].map({'male': 1,'female': 0})
data1
```

```python
from transformers import GPT2Tokenizer, GPT2Model
import torch

# Load pre-trained GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2Model.from_pretrained("gpt2")
```

```python
def generate_text_embeddings(text):
  if len(text) > 0:
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model(**inputs)
    last_hidden_state = outputs.last_hidden_state
    return torch.mean(last_hidden_state, dim=1).detach().numpy()
  else:
    return None
```

```python
data1['ctext1'] = data1['ctext'].apply(generate_text_embeddings)
```

```python
data1['ctweetdesc1'] = data1['ctweetdesc'].apply(generate_text_embeddings)
```

```python
data1['cdescription1'] = data1['cdescription'].apply(generate_text_embeddings)
```

```python
data1.dropna(subset=["cdescription1"], inplace=True)
data1
```

```python
from sklearn.model_selection import train_test_split

X_traina, X_testa, y_traina, y_testa = train_test_split(np.vstack(data1['ctext1']), data1.numgender, test_size=0.35 )
X_trainb, X_testb, y_trainb, y_testb = train_test_split(np.vstack(data1['ctweetdesc1']), data1.numgender, test_size=0.35)
X_trainc, X_testc, y_trainc, y_testc = train_test_split(np.vstack(data1['cdescription1']), data1.numgender, test_size=0.35 )
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_traina, y_traina)
rf.fit(X_trainb, y_trainb)
rf.fit(X_trainc, y_trainc)
```

```python
y_pred1 = rf.predict(X_testa)
y_pred2 = rf.predict(X_testb)
y_pred3 = rf.predict(X_testc)
```

```python
print(accuracy_score(y_testa, y_pred1))
print(accuracy_score(y_testb, y_pred2))
print(accuracy_score(y_testc, y_pred3))
```
