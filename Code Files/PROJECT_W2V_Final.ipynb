## Preprocessing

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

# Define the preprocessing function
def preprocess_text(text):
    # 1. Convert text to lowercase
    text = text.lower()

    # 2. Remove special, non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)

    # 3. Remove stopwords using NLTK
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    text = ' '.join([word for word in word_tokens if word not in stop_words])

    # 4. Remove emoticons, retweets, favorites, hashtags, URLs, and usernames starting with '@'
    text = re.sub(r'@\w+\s?', '', text)  # Remove usernames starting with '@'
    text = re.sub(r'@(\w+)', '', text)
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'#\w+', '', text)  # Remove hashtags
    text = re.sub(r'RT\s', '', text)  # Remove retweets
    text = re.sub(r'&\w+;', '', text)  # Remove HTML entities (if any)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 5. Remove duplicates
    text = ' '.join(dict.fromkeys(text.split()))
    # 6. Tokenize using NLTK
    tokens = word_tokenize(text)

    return ' '.join(tokens)
```

# **W2V**

```python
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
```

```python
def preprocess_text1(text):
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    return tokens
```

```python
data1['ctext1'] = data1['ctext'].apply(preprocess_text1)
```

```python
data1['ctweetdesc1'] = data1['ctweetdesc'].apply(preprocess_text1)
```

```python
data1['cdescription1'] = data1['cdescription'].apply(preprocess_text1)
```

```python
word2vec_model1 = Word2Vec(sentences=data1['ctext1'], vector_size=100, window=5, min_count=1, workers=4)
word2vec_model2 = Word2Vec(sentences=data1['ctweetdesc1'], vector_size=100, window=5, min_count=1, workers=4)
word2vec_model3 = Word2Vec(sentences=data1['cdescription1'], vector_size=100, window=5, min_count=1, workers=4)
```

```python
def average_word_vectors(words, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,), dtype="float64")
    nwords = 0.

    for word in words:
        if word in vocabulary:
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model.wv[word])

    if nwords:
        feature_vector = np.divide(feature_vector, nwords)

    return feature_vector
```

```python
import numpy as np
X1 = np.array([average_word_vectors(words, word2vec_model1, word2vec_model1.wv.index_to_key, 100) for words in data1['ctext1']])
y1 = np.array(data1['numgender'])

X2 = np.array([average_word_vectors(words, word2vec_model2, word2vec_model2.wv.index_to_key, 100) for words in data1['ctweetdesc1']])
y2 = np.array(data1['numgender'])

X3 = np.array([average_word_vectors(words, word2vec_model3, word2vec_model3.wv.index_to_key, 100) for words in data1['cdescription1']])
y3 = np.array(data1['numgender'])
```

```python
from sklearn.model_selection import train_test_split

X_train1, X_test1, y_train1, y_test1 = train_test_split(data.ctext, data.numgender, test_size=0.35,random_state=42 )
X_train2, X_test2, y_train2, y_test2 = train_test_split(data.ctweetdesc, data.numgender, test_size=0.35,random_state=42 )
X_train3, X_test3, y_train3, y_test3 = train_test_split(data.cdescription, data.numgender, test_size=0.35,random_state=42 )

X_traina, X_testa, y_traina, y_testa = train_test_split(X1,y1, test_size=0.40)
X_trainb, X_testb, y_trainb, y_testb = train_test_split(X2,y2, test_size=0.40)
X_trainc, X_testc, y_trainc, y_testc = train_test_split(X3,y3, test_size=0.40 )
```
